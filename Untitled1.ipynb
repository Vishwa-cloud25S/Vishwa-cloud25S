```python
import cv2
import torch
import numpy as np

def load_video_64(path):
    cap = cv2.VideoCapture(path)
    frames = []

    while True:
        ret, frame = cap.read()
        if not ret:
            break
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        frame = cv2.resize(frame, (224, 224))
        frames.append(frame)

    cap.release()

    if len(frames) == 0:
        return None

    # Uniform sample to 64 frames
    indices = np.linspace(0, len(frames) - 1, 64).astype(int)
    sampled = [frames[i] for i in indices]

    video = np.stack(sampled).astype(np.float32) / 255.0

    # Kinetics normalization
    mean = np.array([0.45, 0.45, 0.45])
    std = np.array([0.225, 0.225, 0.225])
    video = (video - mean) / std

    video = torch.tensor(video).permute(3, 0, 1, 2)  # (3,64,224,224)

    return video
```

```python
correct_consistency = 0
total = 0
confidence_drops = []

for cls in SELECTED_CLASSES:
    class_folder = os.path.join("/content/UCF5_subset", cls)

    for vid in os.listdir(class_folder):
        video_path = os.path.join(class_folder, vid)

        # CLEAN
        clean_tensor = load_video_64(video_path)
        if clean_tensor is None:
            continue

        clean_tensor = clean_tensor.unsqueeze(0).to(device)

        with torch.no_grad():
            clean_output = i3d(clean_tensor)
            clean_probs = F.softmax(clean_output, dim=1)
            clean_conf, clean_pred = torch.max(clean_probs, dim=1)

        # BLUR
        blur_tensor = load_video_blur_sigma4(video_path)
        blur_tensor = blur_tensor.unsqueeze(0).to(device)

        with torch.no_grad():
            blur_output = i3d(blur_tensor)
            blur_probs = F.softmax(blur_output, dim=1)
            blur_conf, blur_pred = torch.max(blur_probs, dim=1)

        # Consistency check
        if clean_pred.item() == blur_pred.item():
            correct_consistency += 1

        # Confidence drop
        confidence_drops.append(clean_conf.item() - blur_conf.item())

        total += 1

print("======================================")
print("Total videos processed:", total)
print("Prediction Consistency (%):", (correct_consistency / total) * 100)
print("Average Confidence Drop:", sum(confidence_drops) / len(confidence_drops))
print("======================================")
```

```python
correct_consistency_ts = 0
total_ts = 0
confidence_drops_ts = []

for cls in SELECTED_CLASSES:
    class_folder = os.path.join("/content/UCF5_subset", cls)

    for vid in os.listdir(class_folder):
        video_path = os.path.join(class_folder, vid)

        # CLEAN
        clean_tensor = load_video_64(video_path)
        if clean_tensor is None:
            continue

        clean_tensor = clean_tensor.unsqueeze(0).to(device)
        clean_tensor_ts = clean_tensor.permute(0,2,1,3,4)  # B,T,C,H,W

        with torch.no_grad():
            clean_output = timesformer(pixel_values=clean_tensor_ts)
            clean_probs = F.softmax(clean_output.logits, dim=1)
            clean_conf, clean_pred = torch.max(clean_probs, dim=1)

        # BLUR
        blur_tensor = load_video_blur_sigma4(video_path)
        blur_tensor = blur_tensor.unsqueeze(0).to(device)
        blur_tensor_ts = blur_tensor.permute(0,2,1,3,4)

        with torch.no_grad():
            blur_output = timesformer(pixel_values=blur_tensor_ts)
            blur_probs = F.softmax(blur_output.logits, dim=1)
            blur_conf, blur_pred = torch.max(blur_probs, dim=1)

        if clean_pred.item() == blur_pred.item():
            correct_consistency_ts += 1

        confidence_drops_ts.append(clean_conf.item() - blur_conf.item())

        total_ts += 1

print("======================================")
print("TimeSformer Results")
print("Total videos:", total_ts)
print("Prediction Consistency (%):", (correct_consistency_ts / total_ts) * 100)
print("Average Confidence Drop:", sum(confidence_drops_ts)/len(confidence_drops_ts))
print("======================================")
```

```python
import torch.nn as nn

# Replace final projection layer for 5 classes
in_features = i3d.blocks[-1].proj.in_features
i3d.blocks[-1].proj = nn.Linear(in_features, len(SELECTED_CLASSES))

i3d = i3d.to(device)

print("I3D head modified for", len(SELECTED_CLASSES), "classes")
```

```python
from torch.utils.data import Dataset, DataLoader

class UCF5Dataset(Dataset):
    def __init__(self, root_dir):
        self.samples = []
        self.root_dir = root_dir

        for label, cls in enumerate(SELECTED_CLASSES):
            class_folder = os.path.join(root_dir, cls)
            for vid in os.listdir(class_folder):
                self.samples.append(
                    (os.path.join(class_folder, vid), label)
                )

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        video_path, label = self.samples[idx]
        video_tensor = load_video_64(video_path)
        return video_tensor, torch.tensor(label)
```

```python
from sklearn.model_selection import train_test_split

dataset = UCF5Dataset("/content/UCF5_subset")

train_indices, test_indices = train_test_split(
    list(range(len(dataset))),
    test_size=0.2,
    random_state=42,
    stratify=[label for _, label in dataset.samples]
)

train_subset = torch.utils.data.Subset(dataset, train_indices)
test_subset = torch.utils.data.Subset(dataset, test_indices)

train_loader = DataLoader(train_subset, batch_size=2, shuffle=True)
test_loader = DataLoader(test_subset, batch_size=2, shuffle=False)

print("Train size:", len(train_subset))
print("Test size:", len(test_subset))
```

```python
import torch.optim as optim

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(i3d.parameters(), lr=1e-4)
```

```python
epochs = 3

for epoch in range(epochs):
    i3d.train()
    total_loss = 0

    for videos, labels in train_loader:
        videos = videos.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()

        outputs = i3d(videos)
        loss = criterion(outputs, labels)

        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    avg_loss = total_loss / len(train_loader)
    print(f"Epoch {epoch+1}/{epochs}, Loss: {avg_loss:.4f}")
```

```python
def load_video_dropout_40_32(path):
    cap = cv2.VideoCapture(path)
    frames = []

    while True:
        ret, frame = cap.read()
        if not ret:
            break
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        frame = cv2.resize(frame, (224, 224))
        frames.append(frame)

    cap.release()

    if len(frames) == 0:
        return None

    # First sample to 32 frames
    indices = np.linspace(0, len(frames) - 1, 32).astype(int)
    sampled = [frames[i] for i in indices]

    # Apply 40% dropout
    keep_count = int(32 * 0.6)  # keep 60%
    keep_indices = sorted(np.random.choice(range(32), keep_count, replace=False))
    kept_frames = [sampled[i] for i in keep_indices]

    # Pad back to 32 frames
    while len(kept_frames) < 32:
        kept_frames.append(kept_frames[-1])

    video = np.stack(kept_frames).astype(np.float32) / 255.0

    mean = np.array([0.45, 0.45, 0.45], dtype=np.float32)
    std = np.array([0.225, 0.225, 0.225], dtype=np.float32)
    video = (video - mean) / std

    video = torch.tensor(video, dtype=torch.float32)
    video = video.permute(3, 0, 1, 2)

    return video
```

```python
i3d.eval()

correct_dropout = 0
total_dropout = 0

with torch.no_grad():
    for idx in test_indices:
        video_path, label = dataset.samples[idx]

        video_tensor = load_video_dropout_40_32(video_path)
        if video_tensor is None:
            continue

        video_tensor = video_tensor.unsqueeze(0).to(device)
        label = torch.tensor(label).unsqueeze(0).to(device)

        outputs = i3d(video_tensor)
        pred = torch.argmax(outputs, dim=1)

        correct_dropout += (pred == label).sum().item()
        total_dropout += 1

dropout_accuracy = correct_dropout / total_dropout * 100

print("Dropout Correct:", correct_dropout)
print("Dropout Total:", total_dropout)
print("Dropout Accuracy:", dropout_accuracy)
```

Highly sensitive to spatial blur.

Not very sensitive to temporal dropout (in this small test).
I3D may rely more on spatial clarity than temporal continuity in this small setup.

```python
def load_video_shuffle(path, window=4):
    cap = cv2.VideoCapture(path)
    frames = []

    while True:
        ret, frame = cap.read()
        if not ret:
            break
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        frame = cv2.resize(frame, (224, 224))
        frames.append(frame)

    cap.release()

    indices = np.linspace(0, len(frames)-1, 32).astype(int)
    sampled = [frames[i] for i in indices]

    # shuffle in small temporal windows
    for i in range(0, len(sampled), window):
        chunk = sampled[i:i+window]
        np.random.shuffle(chunk)
        sampled[i:i+window] = chunk

    video = np.stack(sampled).astype(np.float32) / 255.0
    video = torch.tensor(video).permute(3,0,1,2)
    return video
```

```python
def load_video_hybrid(path):
    cap = cv2.VideoCapture(path)
    frames = []

    while True:
        ret, frame = cap.read()
        if not ret:
            break
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        frame = cv2.resize(frame, (224, 224))
        frame = cv2.GaussianBlur(frame, (9,9), 4)
        frames.append(frame)

    cap.release()

    indices = np.linspace(0, len(frames)-1, 32).astype(int)
    sampled = [frames[i] for i in indices]

    # Drop 40%
    keep_count = int(len(sampled) * 0.6)
    sampled = sampled[:keep_count]

    while len(sampled) < 32:
        sampled.append(sampled[-1])

    video = np.stack(sampled).astype(np.float32) / 255.0
    video = torch.tensor(video).permute(3,0,1,2)
    return video
```

```python
import torch
import torch.nn.functional as F
import numpy as np
import matplotlib.pyplot as plt
import cv2

def generate_gradcam_i3d(model, video_tensor, title="GradCAM"):
    model.eval()

    features = []
    gradients = []

    # Target last convolution block before classifier
    target_layer = model.blocks[-2]

    def forward_hook(module, input, output):
        features.append(output)

    def backward_hook(module, grad_input, grad_output):
        gradients.append(grad_output[0])

    handle_f = target_layer.register_forward_hook(forward_hook)
    handle_b = target_layer.register_full_backward_hook(backward_hook)

    video_tensor = video_tensor.unsqueeze(0).to(device)
    video_tensor.requires_grad = True

    output = model(video_tensor)
    pred_class = torch.argmax(output, dim=1)

    model.zero_grad()
    output[0, pred_class].backward()

    feature_map = features[0]
    gradient = gradients[0]

    weights = torch.mean(gradient, dim=(2,3,4), keepdim=True)
    cam = torch.sum(weights * feature_map, dim=1)
    cam = F.relu(cam)

    cam = cam.squeeze().detach().cpu().numpy()
    cam = (cam - cam.min()) / (cam.max() - cam.min() + 1e-8)

    # Take middle frame heatmap
    cam_frame = cam[cam.shape[0] // 2]

    # Original middle frame
    original = video_tensor[0,:,video_tensor.shape[2]//2].detach().cpu().permute(1,2,0).numpy()

    heatmap = cv2.resize(cam_frame, (224,224))
    heatmap = np.uint8(255 * heatmap)
    heatmap = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)
    heatmap = heatmap / 255.0

    overlay = 0.6 * original + 0.4 * heatmap

    plt.figure(figsize=(6,6))
    plt.imshow(overlay)
    plt.title(title)
    plt.axis("off")
    plt.show()

    handle_f.remove()
    handle_b.remove()
```
